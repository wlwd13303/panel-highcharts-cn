```python
import panel_highcharts as ph

ph.config.theme("auto")
```

```python
import panel as pn
pn.extension('highchart')
```

```python
import urllib, requests
system_proxies = urllib.request.getproxies()


if system_proxies and 'http' in system_proxies:
    data = requests.get('https://data.jianshukeji.com/stock/history/000001', proxies={"https": system_proxies['http']})
else:
    data = requests.get('https://data.jianshukeji.com/stock/history/000001')

data.json()['data']
```

```python
configuration = {
    "title": {"text": ""},
    "rangeSelector": {
        "buttons": [
            {"type": "year", "count": 1, "text": "1"},
            {"type": "month", "count": 1, "text": "1"},
            {"type": "week", "count": 1, "text": "1"},
            {"type": "all", "count": 1, "text": ""},
        ],
        "selected": 1,
        "inputEnabled": False,
    },
    "plotOptions": {
            "series": {
                "showInLegend": True
            }
        },
    "series": [
        {"name": "", "id": '000001', "type": "line", "data": data.json()['data'], "tooltip": {"valueDecimals": 2}}
    ],
    "tooltip": {
        "split": False,
        "shared": True
    },
}
```

```python
chart = ph.HighStock(object=configuration, sizing_mode="stretch_both", min_height=600)
chart
```
